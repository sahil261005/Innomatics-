```python
# Problem 1: Social Media – Post Engagement Analyzer

def analyze_engagement(likes_list):
    """Calculates total likes and determines if a post is viral."""
    total_likes = 0
    for like in likes_list:
        total_likes += like

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print(f"Total Likes: {total_likes}")
    print(f"Post Status: {status}")

# Execution
analyze_engagement([400, 300, 350])
```

```python
# Problem 2: Healthcare – Medicine Stock Alert System

def monitor_medicine_stock(stock):
    """Checks medicine inventory levels and flags low stock."""
    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"

    print(f"Medicine Stock: {stock}")
    print(f"Status: {status}")

# Execution
monitor_medicine_stock(6)
```

```python
# Problem 3: Agriculture – Rainfall Adequacy Checker

def check_rainfall_adequacy(rainfall_data, required_level):
    """Analyzes seasonal rainfall to check if it meets the required average."""
    total_rainfall = sum(rainfall_data)
    average_rainfall = total_rainfall / len(rainfall_data)

    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print(f"Average Rainfall: {average_rainfall:.0f}")
    print(f"Rainfall Status: {status}")

# Execution
check_rainfall_adequacy([80, 60, 75, 73], 70)
```

```python
# Problem 4: Social Media – Duplicate Account Detection

def detect_duplicate_accounts(usernames):
    """Detects if duplicate usernames exist using Set data structure."""
    if len(usernames) != len(set(usernames)):
        result = "Yes"
    else:
        result = "No"

    print(f"Duplicate Accounts Found: {result}")

# Execution
usernames_list = ["alice_99", "bob_dev", "alice_99"]
detect_duplicate_accounts(usernames_list)
```

```python
# Problem 5: Healthcare – Appointment Eligibility Checker

def check_appointment_eligibility(age):
    """Validates patient age for appointment eligibility."""
    if age >= 18:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print(f"Patient Age: {age}")
    print(f"Eligibility Status: {status}")

# Execution
check_appointment_eligibility(21)
```

```python
# Problem 6: Agriculture – Premium Crop Price Filter

def filter_premium_crops(crop_prices):
    """Filters and displays prices that exceed the premium threshold."""
    premium_threshold = 2000
    premium_crops = [price for price in crop_prices if price > premium_threshold]

    print(f"Premium Crops: {premium_crops}")

# Execution
filter_premium_crops([1500, 2500, 3200, 1800])
```

```python
# Problem 7: System Monitoring – Application Health Checker

def evaluate_app_health(error_count):
    """Evaluates system status based on total error count."""
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print(f"Error Count: {error_count}")
    print(f"System Status: {status}")

# Execution
evaluate_app_health(7)
```

```python
# Problem 8: Banking – Daily Transaction Limit Checker

def validate_transaction_limit(amount):
    """Checks if a transaction amount is within the daily limit of ₹50,000."""
    daily_limit = 50000
    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"

    print(f"Transaction Amount: {amount}")
    print(f"Transaction Status: {status}")

# Execution
validate_transaction_limit(60000)
```

```python
# Problem 9: E-Learning – Student Attendance Eligibility System

def check_exam_eligibility(attendance_record):
    """Calculates attendance percentage and verifies exam eligibility."""
    present_count = attendance_record.count("P")
    total_days = len(attendance_record)
    percentage = (present_count / total_days) * 100

    if percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"

    print(f"Attendance Percentage: {percentage:.1f}")
    print(f"Exam Eligibility: {eligibility}")

# Execution
attendance = ["P", "P", "A", "P", "P"]
check_exam_eligibility(attendance)
```

```python
# Problem 10: Smart Electricity Bill Analyzer

def smart_electricity_analyzer(units):
    """Calculates electricity bill based on units and classifies usage."""
    # Calculating Bill Amount
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    # Classification logic
    if bill < 500:
        usage_level = "Low Usage"
    elif 500 <= bill <= 1500:
        usage_level = "Moderate Usage"
    else:
        usage_level = "High Usage"

    print(f"Units Consumed: {units}")
    print(f"Total Bill: ₹{bill}")
    print(f"Usage Status: {usage_level}")
    return bill, usage_level

# Execution
smart_electricity_analyzer(250)
```
