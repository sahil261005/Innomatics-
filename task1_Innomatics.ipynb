```python
#2
marks = [45, 78, 90, 33, 60]
pass_count = 0
fail_count = 0


for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

print(f"Total Pass Students: {pass_count}")
print(f"Total Fail Students: {fail_count}")
```

```python
# 3
names = [" Alice ", "bob", " CHARLIE "]


cleaned_names = [name.strip().lower() for name in names]

print(f"Cleaned List: {cleaned_names}")
```

```python
# 4
messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    length = len(msg)

    flag = " [LONG MESSAGE]" if length > 10 else ""
    print(f"Length: {length} and Message: {msg}{flag}")
```
