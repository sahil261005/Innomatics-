# **Problem 1:Smart Parking Lot Management System**

```python
def parking_system(capacity, vehicle_logs):
    current = 0
    peak = 0

    for log in vehicle_logs:
        if log == "IN":
            current += 1
        elif log == "OUT":
            if current > 0:
                current -= 1

        if current > peak:
            peak = current

    print("Currently Parked Vehicles:", current)

    if current > capacity:
        print("Parking Status: Exceeded Capacity")
    else:
        print("Parking Status: Available")


capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

parking_system(capacity, logs)
```

# **Problem 2:Online Food Delivery Time Estimator**

```python
def delivery_time_estimator(distance, weather_condition, traffic):
    base_time = 20
    estimated_time = base_time

   # Distance delay
    if distance <= 10:
        estimated_time += 5
    elif distance <= 20:
        estimated_time += 10
    elif distance <= 30:
        estimated_time += 15

    # Weather delay
    if weather_condition == "Rainy":
        estimated_time += 10
    elif weather_condition == "Stormy":
        estimated_time += 15

    # Traffic delay
    if traffic == "High":
        estimated_time += 15
    elif traffic == "Medium":
        estimated_time += 10
    elif traffic == "Low":
        estimated_time += 5
    print("Estimated Delivery Time:", estimated_time, "minutes")


# Sample Input
delivery_time_estimator(22, "Stormy", "Medium")
```

# **Problem 3:Movie Theatre Seat Occupancy**

```python
def theatre_occupancy(total_seats, booked_seats):
    booked_count = 0

    for seat in booked_seats:
        if seat == 1:
            booked_count += 1


    occupancy = (booked_count / total_seats) * 100
    print("Occupancy:", int(occupancy), "%")

    if occupancy == 100:
        print("Show Status: Housefull")
        print("Suggestion: Open Additional Show")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")


total_seats = 200
booked_seats = [1] * 150

theatre_occupancy(total_seats, booked_seats)
```

# **Problem 4:Cloud Server Load Classification System**

```python
def Server_load(cpu_readings):
  total=0
  for cpu_reading in cpu_readings:
        total += cpu_reading

  average = total / len(cpu_readings)
  if average<=50:
    return "Normal"
  elif average>=51 and average<=80:
    return "Worning"
  else:
      return "Critical"

cpu_readings = [45, 60, 70, 85, 90]
Server_Status = Server_load(cpu_readings)

print("Average:", sum(cpu_readings)/len(cpu_readings))
print("Server_Status:", Server_Status)
```

# **Problem 5:Smart Classroom Resource Usage Monitor**

```python
def classroom_monitor(resources):
    overused = []

    for resource, hours in resources.items():
        if hours > 8:
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")



usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

classroom_monitor(usage)
```

#**Problem 6: Online Event Registration Capacity Controller**

```python
def event_registration(capacity, total_registrations):
    confirmed = 0
    waitlisted = 0

    for i in range(total_registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)

    if confirmed == capacity:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")

capacity = 100
registrations = 105

event_registration(capacity, registrations)
```
