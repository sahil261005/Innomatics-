```
from langchain_core.prompts import PromptTemplate
from langchain_core.prompts import ChatPromptTemplate
```

```
#Replace Hardcoded Prompts
template_1 = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners"
)

# Test
print(template_1.format(topic="Artificial Intelligence"))
```

```
#Multi-Input Prompt System
multi_input_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

# Test Cases
print(multi_input_template.format(topic="AI", audience="beginners", tone="friendly"))
print(multi_input_template.format(topic="Python", audience="kids", tone="fun"))
print(multi_input_template.format(topic="Deep Learning", audience="engineers", tone="technical"))
```

```
#Prompt Variations Engine
# Teaching Template
teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

# Interview Template
interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 interview questions about {topic}"
)

# Storytelling Template
story_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

# Run for topic
topic = "Machine Learning"

print(teaching_template.format(topic=topic))
print(interview_template.format(topic=topic))
print(story_template.format(topic=topic))
```

```
#ChatPromptTemplate System
def get_chat_prompt(role, topic):

    if role == "teacher":
        system_message = "You are a helpful teacher who explains concepts clearly."

    elif role == "interviewer":
        system_message = "You are an interviewer asking challenging questions."

    elif role == "motivator":
        system_message = "You are a motivational speaker who inspires learning."

    else:
        system_message = "You are a helpful assistant."

    chat_template = ChatPromptTemplate.from_messages([
        ("system", system_message),
        ("user", "Tell me about {topic}")
    ])

    return chat_template.format_messages(topic=topic)


# Test
messages = get_chat_prompt("teacher", "Neural Networks")

for msg in messages:
    print(msg)
```

```
#Input Validation Layer
def validate_inputs(audience, tone):

    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        print("Invalid audience! Setting default = beginner")
        audience = "beginner"

    if tone not in valid_tone:
        print("Invalid tone! Setting default = formal")
        tone = "formal"

    return audience, tone


# Test
audience, tone = validate_inputs("kids", "friendly")
print(audience, tone)
```

```
#Prompt Generator App
def generate_prompt(topic, audience, tone, style):

    audience, tone = validate_inputs(audience, tone)

    if style == "teaching":
        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} step by step for {audience} in a {tone} tone"
        )

    elif style == "interview":
        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Ask 3 interview questions about {topic} for {audience} in a {tone} tone"
        )

    elif style == "storytelling":
        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} as a story for {audience} in a {tone} tone"
        )

    else:
        return "Invalid style"

    return template.format(topic=topic, audience=audience, tone=tone)


# Example
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

```
#Template Reusability Test
reusable_template = PromptTemplate(
    input_variables=["topic", "audience"],
    template="Explain {topic} for {audience}"
)

# 5 Different Inputs
inputs = [
    ("AI", "beginners"),
    ("Python", "students"),
    ("Blockchain", "developers"),
    ("Data Science", "analysts"),
    ("Cyber Security", "experts")
]

for topic, audience in inputs:
    print(reusable_template.format(topic=topic, audience=audience))
```

```
#Final Pipeline Flow
def full_pipeline(topic, audience, tone, style):

    # Step 1: Validate
    audience, tone = validate_inputs(audience, tone)

    # Step 2: Generate Prompt
    prompt = generate_prompt(topic, audience, tone, style)

    return prompt


# Test
print(full_pipeline("AI", "beginner", "fun", "teaching"))
```
