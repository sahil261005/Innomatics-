**Task 1: Conceptual Understanding**

1.What is the difference between "Love" and "love" in NLP?

In NLP the difference between "Love" and "love" lies in case sensitivity(capitalization) and how the text is processed.
“Love” is the capitalized form, while “love” is the lowercase form. In a case-sensitive NLP system, these two are treated as different tokens (words) because computers distinguish between uppercase and lowercase letters. For example, in the sentences “I Love NLP” and “I love NLP,” the words “Love” and “love” would be counted separately.

2.What happens if stopwords are not removed?

If stopwords are not removed in NLP, the following happens:

->Text becomes larger and noisy:
Common words like the, is, and, of remain, adding little useful meaning.

-> Higher computational cost:
More words increase memory usage, processing time, and feature size.

-> Important words may get less importance:
Frequent stopwords can dominate word counts and hide meaningful terms.

-> Vocabulary size increases:
The model must handle many extra words, making analysis more complex.

-> Sentence meaning is fully preserved:
Keeping stopwords maintains grammar and context of the original text.

-> May be useful for some NLP tasks:
Tasks like sentiment analysis or translation benefit from keeping stopwords (e.g., “not” changes meaning).

4.Explain the difference between stemming and lemmatization.

Stemming:
It removes prefixes or suffixes from a word using simple rules.
It does not consider the actual meaning or grammar of the word.
The result may not be a valid dictionary word.

Examples:
playing → play,
studies → studi ❌,
running → run, Fast and simple, but less accurate.

Lemmatization:
It converts a word to its base or dictionary form (lemma).
It considers the context and part of speech (POS).
The result is always a meaningful word.

Examples:
playing → play,
studies → study ✅,
better → good,
running → run,
Slower but more accurate.

**Task 3: Stress Testing**

```python
import re

def preprocess_text(text):
    original = text

    # 1. Lowercase
    text = text.lower()

    # 2. Remove URLs & emails
    text = re.sub(r'http\S+|www\S+|\S+@\S+', '', text)

    # 3. Remove numbers
    text = re.sub(r'\d+', '', text)

    # 4. Remove repeated characters (e.g., loooove → loove)
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)

    # 5. Remove symbols (keep only letters & spaces)
    text = re.sub(r'[^a-z\s]', '', text)

    # 6. Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # 7. Remove short words (≤2) but keep "no", "not"
    keep_words = ["no", "not"]
    tokens = [w for w in text.split() if len(w) > 2 or w in keep_words]

    clean_sentence = " ".join(tokens)

    return original, tokens, clean_sentence


# Sample Inputs (your assignment)
sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]


#  Output Format
for s in sentences:
    original, tokens, clean = preprocess_text(s)

    print("Original Text   :", original)
    print("Cleaned Tokens  :", tokens)
    print("Cleaned Sentence:", clean)
    print("-" * 50)
```

**Task 6: Build Full Pipeline**

```python
import re

def clean_text(sentence):
    # Lowercase
    sentence = sentence.lower()
    # Remove punctuation
    sentence = re.sub(r'[^\w\s]', '', sentence)
    return sentence

def tokenize(sentence):
    return sentence.split()

def full_pipeline(text_list):
    clean_sentences = []
    all_tokens = []

    for sentence in text_list:
        cleaned = clean_text(sentence)
        clean_sentences.append(cleaned)

        tokens = tokenize(cleaned)
        all_tokens.extend(tokens)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }


# Example input
text_list = [
    "AI is the future!",
    "Data Science is amazing.",
    "AI & ML are trending."
]

output = full_pipeline(text_list)

print(output)
```
